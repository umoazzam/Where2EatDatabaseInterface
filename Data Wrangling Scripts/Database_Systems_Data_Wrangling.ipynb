```python
groups = pd.read_csv(r"C:\Users\sohan\OneDrive\Desktop\groups.csv")
groups
```

```python
members = groups['members']
members[0]
```

```python
from ast import literal_eval
mid_longitude = []
mid_latitude = []
literal_eval(members[0])
for i in members:
    latitudes = []
    longitudes = []
    for j in literal_eval(i):
        latitudes.append(float(Users.loc[Users['userID'] == j]['latitude']))
        longitudes.append(float(Users.loc[Users['userID'] == j]['longitude']))
    mid_longitude.append(sum(longitudes)/len(longitudes))
    mid_latitude.append(sum(latitudes)/len(latitudes))
mid_latitude
```

```python
groups['midLong'] = mid_longitude
groups['midLat'] = mid_latitude
groups
```

```python
groups.to_csv(r"C:\Users\sohan\OneDrive\Desktop\groups.csv")
```
